```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Normalization= Min max scaler=(x-xmin)/(xmax-xmin)= 0 to 1 values shifts
# Standardization = Standard Scaler=z score=z=(xi-u)/Sigma,Mean 0 std deviation=1 for normal distribution
# Robust Scaler
```

```python
df=pd.read_csv("Social_Network_Ads.csv")
```

```python
df
```

```python
x= df[['Age','EstimatedSalary']]
```

```python
x
```

```python
x.describe()
```

```python
plt.figure(figsize=(15,9))
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title("Age Vs Estimated Salary")
plt.scatter(df['Age'],df['EstimatedSalary'],color='Magenta');
plt.grid();
```

# Min Max Scaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#create an object
scaler =MinMaxScaler()
```

```python
x_scaled=scaler.fit_transform(x)
```

```python
x_scaled
```

```python
x_scaled.max()
```

```python
x_scaled.min()
```

```python
pd.DataFrame(x_scaled).describe()
```

```python
plt.figure(figsize=(15,7))
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.title('Age Vs Estimated Salary')
plt.scatter(x_scaled[:,0],x_scaled[:,1])
plt.grid();
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
Scaler=StandardScaler()
```

```python
y_scaled=Scaler.fit_transform(x)
```

```python
y_scaled
```

```python
pd.DataFrame(y_scaled).describe()
```

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title("Age Vs Sal")
plt.scatter(y_scaled[:,0],y_scaled[:,1])
```

```python
plt.figure(figsize=(15,7))
plt.subplot(2,2,1)
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.title('Age Vs Estimated Salary')
plt.scatter(x_scaled[:,0],x_scaled[:,1],color='magenta')
plt.grid();
plt.subplot(2,2,2)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title("Age Vs Sal")
plt.scatter(y_scaled[:,0],y_scaled[:,1])
plt.grid();
```
