```python
def calc(num):
    sq=num**2
    cu=num**3
    return sq,cu
```

### Read number from user write and call a function to find factorial of it

```python
def fact(n):
    fact=1
    for i in range(1,n+1):
        fact=fact*i
    return fact
```

```python
n=int(input("Enter the number:"))
fact(n)
```
